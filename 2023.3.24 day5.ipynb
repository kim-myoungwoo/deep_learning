```python
# 2010년부터 2016년까지의 부동산 거래 데이터
import pandas as pd

df= pd.read_csv('./data/data/house_train.csv')
df.head()
```

```python
df.dtypes
# df.info()
```

```python
# 결측값이 있는지 확인
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
# 카테고리 데이터 확인
df.MSZoning.value_counts()
```

```python
# get_dummies() 함수를 활용하여  -> one-hot-encoding
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
# 결측값 처리
df = df.fillna(df.mean())

df.isnull().sum().sort_values(ascending=False).head()
```

```python
# 결측값이 1000개 이상인 컬럼 제거 후 결측값 처리
df1= pd.read_csv('./data/data/house_train.csv')
df1.dropna(thresh=700,axis=1,inplace=True)    #  thresh -> NAN이 아닌 값이 개수가 700개 미만인 컬럼 제거
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df1.isnull().sum().sort_values(ascending=False)
```

```python
df1.head()
```

```python
# one-hot-encoding
df1 = pd.get_dummies(df1)

# 결측값 처리
df1.fillna(df.mean(numeric_only=True))

df1.head()
```

```python
# 변수들간의 상관관계
df1_corr = df1.corr()

# SalePrice 칼럼에 대한 상관관계를 확인
df1_corr_sort = df1_corr.sort_values('SalePrice', ascending=False)
df1_corr_sort['SalePrice'].head(10)
```

```python
cols = ['SalePrice','OverallQual','GrLivArea','GarageCars','GarageArea','TotalBsmtSF']

import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df1[cols])
plt.show()
```

```python
# 상관관계가 있는 컬럼만 데이터를 가져옴, target -> SalePrice
cols_train = cols[1:]
X = df1[cols_train]
X.head()
y = df1[cols[1]]

print(X.shape,y.shape)
```

```python
# X와 y를 학습 데이터롸 테스트 데이터로 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
# 모델 생성 -> 학습 , early_stopping, checkpointer  추가해서 모델 실행 -> 검증 데이터는 25%로 설정
#  6 : 2 : 2 -> 학습 : 검증 : 테스트
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
model = Sequential()
model.add(Dense(30, input_dim=X.shape[1], activation='relu')) # 입력층과 첫번쨰 layer 추가
model.add(Dense(20, activation='relu'))  # 두번째층 추가
model.add(Dense(30, activation='relu'))   # 세번째층 추가
model.add(Dense(1))  # 출력층 추가
model.summary()
```

```python
# 컴파일
model.compile(loss='mse', optimizer='adam',metrics=['accuracy'])  #  'loss = mean_squred_error'
# 중단점과 모델 저장 사용  -> 실행시 옵션
early_stopping_callback = EarlyStopping(patience=10)
checkpointer = ModelCheckpoint(filepath='./data/data/model/house.hdf5', verbose=0, save_best_only=True)
model.fit(X_train,y_train, epochs=2000, validation_split=0.25, batch_size=32, callbacks=[early_stopping_callback,checkpointer])
```

```python
from tensorflow.keras.models import load_model

model = load_model('./data/data/model/house.hdf5')
```

```python
model.evaluate(X_test, y_test)
```

```python
model.predict(X_test).flatten()  # model.predict의 결과가 2차원 배열로 반환 -> 1차원으로 평탄화
```

```python
# 모델을 통한 예측값과 실제 값을 비교
real_price =[]
predict_price = []
X_num = []
n =0
y_pred = model.predict(X_test).flatten()
for i in range(25):
    real = y_test.iloc[i]
    pred = y_pred[i]
    print('실제 가격 : {:.2f}   , 예측 가격 : {:2f}'.format(real,pred))
    real_price.append(real)
    predict_price.append(pred)
    n+=1
    X_num.append(n)
```

```python
plt.plot(X_num, real_price, label='real price')
plt.plot(X_num, predict_price, label='predict price')
plt.legend()

plt.show()
```
