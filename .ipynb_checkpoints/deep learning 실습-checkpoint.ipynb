```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import Input, models, layers, optimizers, metrics
from tensorflow.keras.layers import Dense, Flatten, Activation, Dropout
from tensorflow.keras.applications import VGG16, InceptionV3, ResNet50V2
from tensorflow.keras.callbacks import EarlyStopping

import numpy as np
import matplotlib.pyplot as plt
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,   # 정규화
                   horizontal_flip=True, # 수평으로 뒤집기
                   width_shift_range=0.1, # 수평이동
                   height_shift_range=0.1) # 수직이동
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator= train_datagen.flow_from_directory('../archive/lung cancer dataset/lung cancer dataset',   # 학습셋이 있는 폴더의 위치
                                                target_size=(150,150), batch_size=5, class_mode='categorical')

test_generator= test_datagen.flow_from_directory('../archive/lung cancer dataset/lung cancer dataset',   # 학습셋이 있는 폴더의 위치
                                                target_size=(150,150), batch_size=5, class_mode='categorical')
```

```python
transfer_model = ResNet50V2(input_shape=(150,150,3), include_top=False)
transfer_model.trainable =False
transfer_model.summary()
```

```python
model = Sequential()
model.add(transfer_model)
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.2))
# model.add(Dense(32, activation='relu'))
# model.add(Dropout(0.2))
# model.add(Dense(16, activation='relu'))
# model.add(Dropout(0.1))
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
# 컴파일 옵션
model.compile(loss='categorical_crossentropy', optimizer=optimizers.Adam(learning_rate=0.0001), metrics=['accuracy'])

# 조기 중간점 설정
early_stopping = EarlyStopping(patience=5)

# 모델 실행
history = model.fit(train_generator, epochs=30, validation_data=test_generator, validation_steps=10, callbacks=[early_stopping])
```

```python
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len= np.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='.', c='red', label='testset_loss')
plt.plot(x_len, y_loss, marker='.', c='blue', label='trainset.loss')

plt.legend(loc='upper right')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
import os
from PIL import Image

# Define the directory path where your images are stored
directory_path = "../archive/Test cases"

# Load the pre-trained model
# model = history("../archive/Test cases")

# Define the size of the input images expected by the model
img_size = (150,150)
pred_categori_list = []
# Loop through all the files in the directory and make predictions
for filename in os.listdir(directory_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Load the image file using the Pillow libraryL=
        img = Image.open(os.path.join(directory_path, filename))
        # Resize the image to match the input size expected by the model
        img = img.resize(img_size)
        # Convert the image to a numpy array
        img_array = np.array(img)
        # Normalize the pixel values
        img_array = img_array / 255.0
        # Reshape the array to match the expected shape of the input tensor
        img_array = np.reshape(img_array, (1,150,150, 3))
        # Make the prediction using the model
        prediction = model.predict(img_array)
        # Get the label with the highest probability
        label = np.argmax(prediction)
        pred_categori_list.append(label)
        print(f"File: {filename} Predicted Label: {label}")
```

```python
unique_elements, count=np.unique(np.array(pred_categori_list), return_counts=True)

for uniq, cnt in zip(unique_elements, count):
    print(f"{uniq}:{cnt}")
```
