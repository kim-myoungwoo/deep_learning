```python
# 다중 선형 회귀
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y = np.array([81,93,91,97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y)
plt.show()
```

```python
a1, a2, b = (0,0,0)   # 기울기 a1, a2, 절편 b 의 값을 초기화
#y_pred = a1*x1 +a2*x2 +b
lr = 0.01
epochs = 2001
n = len(x1)
```

```python
# 경사 하강법을 시작
for i in range(epochs):
    y_pred = a1*x1 + a2*x2 + b  # 예측값 구현
    error = y - y_pred  # 실제값과 예측값의 차이
    
    a1=a1 -lr*((2/n)*sum(-x1*(error)))
    a2=a2 -lr*((2/n)*sum(-x2*(error)))
    b=b -lr*((2/n)*sum(-error))
    
    if i % 100 ==0:
        print("epochs : {}, 기울기1 : {:3.2f}, 기울기2 : {:3.2f}, 절편 : {}".format(i, a1,a2,b))
        
# 실제 예측된 최종 점수 출력
y_pred = a1*x1 +a2*x2 +b
print("실제 점수 : ", y)
print("예측 점수 : ", y_pred)

# 공부 시간이 5시간이고 과외 횟수가 3인 학생의 점수 예측
print("예측 점수 : ", a1*5 + a2*3 +b)
```

```python
# 텐서플로우와 케라스를 활용한 선현 회귀 분석
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()    # 모델 생성 -> 초기화
model.add(Dense(1, input_dim=1, activation = 'linear'))             # layer 추가
model.summary()    # 모델의 요약 정보 출력
model.compile(optimizer='sgd', loss='mse')   # 모델을 실행할 파라미터 설정

model.fit(x1, y, epochs=2000)    # 모델 학습
```

```python
# 공부 시간이 5시간인 학생의 점수 예측
model.predict([5])
```

```python
plt.scatter(x1,y)
plt.plot(x1, model.predict(x1), 'r')  # model.predict() : 학습한 모델로 예측값 구함
plt.show()
```

```python
hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할 경우의 예상 점수는 %.02f점입니다."% (hour, prediction))
```

```python
# 다중 선형 회귀를 텐서플로에서 실행
x1 = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])

model1 = Sequential()    # 모델 생성 -> 초기화
model1.add(Dense(1, input_dim=2, activation = 'linear'))             # layer 추가
model1.summary()    # 모델의 요약 정보 출력
model1.compile(optimizer='sgd', loss='mse')   # 모델을 실행할 파라미터 설정

model1.fit(x1, y, epochs=2000)    # 모델 학습
```

```python
model1.predict([[7,2]])  # 공부한 시간 : 7, 과외 횟수 :2 인 학생의 예측 점수
```

# chap6 로지스틱 회귀모델

```python
x=np.array([2,4,6,8,10,12,14])
y=np.array([0,0,0,1,1,1,1])

model = Sequential()
model.add(Dense(1, input_dim=1, activation='sigmoid'))  # 활성화 함수 : 시그모이드 함수 사용
model.compile(optimizer='sgd', loss='binary_crossentropy')  # 시그모이드 -> 크로스 엔트로피 함수 사용
model.fit(x,y,epochs=2000)
```

```python
# 그래프로 확인
plt.scatter(x,y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
# 학습 시간 7을 입력했을 때 합격 여부 확인
model.predict([7])
```
